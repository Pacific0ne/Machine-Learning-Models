```python
import ast
def convertgenre(obj):
    gnrlst=[]
    for i in ast.literal_eval(obj):
        gnrlst.append(i['name'])
    return gnrlst
```

```python
movies['genres'] = movies['genres'].apply(convertgenre)
```

```python
movies.head()
```

```python
movies['keywords'] = movies['keywords'].apply(convertgenre)
```

```python
movies.head()
```

```python
def convertcast(obj):
    counter=0
    lst=[]
    for i in ast.literal_eval(obj):
        if counter != 3:
            lst.append(i['name'])
            counter+=1
        else:
            break
    return lst
```

```python
movies['cast'] = movies['cast'].apply(convertcast)
```

```python
movies.head()
```

```python
def fetchdirector(obj):
    lst=[]
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            lst.append(i['name'])
            break
    return lst
```

```python
movies['crew'] = movies['crew'].apply(fetchdirector)
```

```python
movies['genres'] = movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])     #removing the spaces from the words as machine might take it as different tokens
movies['keywords'] = movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movies['cast'] = movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movies['crew'] = movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies.head()
```

```python
movies["tags"] = movies["genres"]+movies["keywords"]+movies["cast"]+movies["crew"]
```

```python
movies.head()
```

```python
new_df = movies[['movie_id','title','tags']]
```

```python
new_df.head()
```

```python
new_df["tags"] = new_df['tags'].apply(lambda x:" ".join(x))  #converting the list into string
```

```python
new_df.head()
```

```python
new_df["tags"][0]  #visualising the first row
```

```python
new_df["tags"] = new_df["tags"].apply(lambda x:x.lower()) #converting the words into lowercase
```

```python
new_df.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer  #creating vectors out of tags columnn for each data
```

```python
import nltk  #natural language tool kit
```

```python
from nltk.stem.porter import PorterStemmer  # it will remove and change the similar words like dance, dancer, dancing into dance
ps =PorterStemmer()
```

```python
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
new_df['tags']=new_df["tags"].apply(stem)
```

```python
new_df.head(4)
```

```python
cv = CountVectorizer(max_features=5000,stop_words='english')   #it removes the stop words, like it ,a ,to, in
vectors =cv.fit_transform(new_df['tags']).toarray()
```

```python
len(cv.get_feature_names())
```

```python
vectors
```

```python
cv.get_feature_names()
```

```python
from sklearn.metrics.pairwise import cosine_similarity  #now we will find angle difference between the data vectors
```

```python
similarity = cosine_similarity(vectors)
```

```python
similarity
```

```python
new_df[new_df["title"]=="Batman"].index[0]
```

```python
sorted(list(enumerate(similarity[0])),reverse=True, key=lambda x:x[1])[1:6] 
#our aim is to find the most similar movies from the given movie name, for which we will emunerate , which will give a tuple of index and value, then we will arrange it in decending order
```

```python
new_df.iloc[2409].title  #we are getting siimilar movies
```

```python
def recommend(movie):        #function to get list of top 5 recommended movies
    if movie in new_df["title"].values:
        movie_index = new_df[new_df["title"]==movie].index[0]
        distances = similarity[movie_index]
        movie_list = sorted(list(enumerate(distances)),reverse=True, key=lambda x:x[1])[1:6]
        for i in movie_list:
            print(new_df.iloc[i[0]].title)
    else:
        print("Movie doesn't exist in database.")
```

```python
recommend("Avatar")
```
